```python
# ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib # 凡例等で日本語表記時に必要
import pandas as pd

# CSVファイルから読込
sales = pd.read_csv('C:\\Users\Owner\Matplotlib-Exercises6\\FILE1.csv')
```

```python
# データフレームの内容
sales
```

```python
# X軸とY軸の値
X = sales['商品名']
Y = sales['売上高']

# グラフ装飾
plt.ylim(0, 6000)
plt.title('各商品売上', fontsize=20)
plt.xlabel('商品', fontsize=13)
plt.ylabel('売上', fontsize=13)
plt.grid(True)
plt.tick_params(labelsize=13)

# グラフ描画
plt.bar(X, Y, tick_label=X, color='g', align='center')
plt.show()
```

```python
# X軸とY軸の値
X = sales['商品名']
Y = sales['売上高']
C = ['r', 'c', 'b', 'm', 'y', 'g']

# グラフ装飾
plt.title('各商品売上', fontsize=20)
plt.rcParams['font.size'] = 12.0

# グラフ描画(凡例付き)
plt.pie(Y, counterclock=True, autopct="%1.1f%%", colors=C)
plt.legend(X, fontsize=12, bbox_to_anchor=(0.95, 0.75))
plt.show()
```

```python
# X軸とY軸の値
X = sales['商品名']
Y = sales['売上高']
C = ['r', 'c', 'b', 'm', 'y', 'g']

# 各グラフの描画領域を指定
plt.figure(figsize=(25, 10))
plt.subplots_adjust(wspace=0.2, hspace=0)

# 棒グラフの描画位置
plt.subplot(1, 2, 1)

# 棒グラフ装飾
plt.ylim(0, 6000)
plt.title('各商品売上', fontsize=20)
plt.xlabel('商品', fontsize=13)
plt.ylabel('売上', fontsize=13)
plt.grid(True)
plt.tick_params(labelsize=13)

# 棒グラフ描画
plt.bar(X, Y, tick_label=X, color='g', align='center')

# 円グラフの描画位置
plt.subplot(1, 2, 2)

# 円グラフ装飾
plt.title('各商品売上', fontsize=20)
plt.rcParams['font.size'] = 12.0

# 円グラフ描画(凡例付き)
plt.pie(Y, counterclock=True, autopct="%1.1f%%", colors=C)
plt.legend(X, fontsize=12, bbox_to_anchor=(0.95, 0.75))

# 棒グラフと円グラフの可視化
plt.show()
```
